```python
import numpy as np
from tqdm.auto import tqdm
from fuzzywuzzy import fuzz

def make_df(pairs, data):
    gc.collect()
    
    df = pairs.copy()
    
    categories1 = data.loc[pairs.variantid1, 'categories']
    categories2 = data.loc[pairs.variantid2, 'categories']
    df['is_eq_category_3'] = categories1.apply(lambda x: x['3']).values == categories2.apply(lambda x: x['3']).values
    df['is_eq_category_4'] = categories1.apply(lambda x: x['4']).values == categories2.apply(lambda x: x['4']).values
    
    name_embeds1 = data.loc[pairs.variantid1, 'name_bert_64'].values
    name_embeds2 = data.loc[pairs.variantid2, 'name_bert_64'].values
    
    dist1, dist2 = [], []
    for i in tqdm(range(len(pairs))):
        dist1.append(
            np.abs(name_embeds1[i] - name_embeds2[i]).sum()
        )
        dist2.append(
            ((name_embeds1[i] - name_embeds2[i])**2).sum()**0.5
        )
    df['name_dist1'] = dist1
    df['name_dist2'] = dist2
    
    main_pic_embeds1 = data.loc[pairs.variantid1, 'main_pic_embeddings_resnet_v1'].apply(lambda x: x[0]).values
    main_pic_embeds2 = data.loc[pairs.variantid2, 'main_pic_embeddings_resnet_v1'].apply(lambda x: x[0]).values
    dist1, dist2 = [], []
    for i in tqdm(range(len(pairs))):
        dist1.append(
            np.abs(main_pic_embeds1[i] - main_pic_embeds2[i]).sum()
        )
        dist2.append(
            ((main_pic_embeds1[i] - main_pic_embeds2[i])**2).sum()**0.5
        )
    df['main_pic_dist1'] = dist1
    df['main_pic_dist2'] = dist2
    
    color1 = data.loc[pairs.variantid1, 'color_parsed'].values
    color2 = data.loc[pairs.variantid2, 'color_parsed'].values
    same_colors = []
    for i in tqdm(range(len(pairs))):
        if color1[i] is None or color2[i] is None:
            same_colors.append(0)
        else:
            same_colors.append(
                len(set(color1[i]) & set(color2[i]))
            )
    df['same_colors'] = same_colors
    
    name1 = data.loc[pairs.variantid1, 'name'].values
    name2 = data.loc[pairs.variantid2, 'name'].values
    dist, partial_dist, token_sort_dist, token_set_dist = [], [], [], []
    for i in tqdm(range(len(pairs))):
        dist.append(
            fuzz.ratio(name1[i], name2[i])
        )
        partial_dist.append(
            fuzz.partial_ratio(name1[i], name2[i])
        )
        token_sort_dist.append(
            fuzz.token_sort_ratio(name1[i], name2[i])
        )
        token_set_dist.append(
            fuzz.token_set_ratio(name1[i], name2[i])
        )
    df['name_dist'] = dist
    df['name_partial_dist'] = partial_dist
    df['name_token_sort_dist'] = token_sort_dist
    df['name_token_set_dist'] = token_set_dist
    
    pic_embeds1 = data.loc[pairs.variantid1, 'pic_embeddings_resnet_v1'].values
    pic_embeds2 = data.loc[pairs.variantid2, 'pic_embeddings_resnet_v1'].values
    pics1, pics2 = [], []
    for i in tqdm(range(len(pairs))):
        sz1 = 0
        if pic_embeds1[i] is not None:
            sz1 = pic_embeds1[i].shape[0]
        sz2 = 0
        if pic_embeds2[i] is not None:
            sz2 = pic_embeds2[i].shape[0]
        pics1.append(sz1)
        pics2.append(sz2)
    df['pics1'] = pics1
    df['pics2'] = pics2
    df['pics_diff'] = np.abs(df['pics1'] - df['pics2'])
    
    attr1 = data.loc[pairs.variantid1, 'characteristic_attributes_mapping'].values
    attr2 = data.loc[pairs.variantid2, 'characteristic_attributes_mapping'].values
    attrs_inter, attrs_union = [], []
    for i in tqdm(range(len(pairs))):
        inter = 0
        for key, value in attr1[i].items():
            inter += attr2[i].get(key, None) == value
        attrs_inter.append(inter)
        attrs_union.append(len(set(attr1[i].keys()) | set(attr2[i].keys())))
    df['attrs_inter'] = attrs_inter
    df['attrs_union'] = attrs_union
    df['attrs_inter/union'] = df['attrs_inter']/df['attrs_union']
    
    df[[f'name_embeds1_{i}' for i in range(64)]] = list(name_embeds1)
    df[[f'name_embeds2_{i}' for i in range(64)]] = list(name_embeds2)
    df[[f'main_pic_embeds1_{i}' for i in range(128)]] = list(main_pic_embeds1)
    df[[f'main_pic_embeds2_{i}' for i in range(128)]] = list(main_pic_embeds2)
    df['name1'] = name1
    df['name2'] = name2
    df['category1'] = categories1.apply(lambda x: x['4']).values
    df['category2'] = categories2.apply(lambda x: x['4']).values
    # df[[f'name_embeds_diff_{i}' for i in range(64)]] = list(name_embeds1 - name_embeds2)
    
    return df.drop(['variantid1', 'variantid2'], axis=1)


train_df = make_df(train_pairs, train_data)
test_df = make_df(test_pairs, test_data)
train_df
```

```python
from catboost import CatBoostClassifier, Pool, cv
```

```python
train_pool = Pool(
    data=train_df.drop('target', axis = 1),
    label=train_df['target'].astype(int),
    text_features=['name1', 'name2', 'category1', 'category2']
)

params = {
    'loss_function': 'CrossEntropy',
    'custom_metric': ['AUC'],
    'task_type': 'CPU',
    'random_seed': 56,
    'max_depth': 7
}

cv_data = cv(
    params = params,
    pool = train_pool,
    fold_count=3,
    shuffle=True,
    partition_random_seed=0,
    plot=True,
    stratified=True,
    verbose=False
)
```

```python
max_iter = cv_data['test-AUC-mean'].argmax()
print(f"Best iteration: {max_iter}\nBest AUC: {round(cv_data.iloc[max_iter]['test-AUC-mean'], 4)}±{round(cv_data.iloc[max_iter]['test-AUC-std'], 4)}")
```
